ejercicio sobre antenas y cuadrantes

````python
import random
import math
import matplotlib.pyplot as plt

# ----------------------------
# Parámetros generales
# ----------------------------
LIMITE = 100
NUM_USUARIOS = random.randint(30, 60)
RADIO_ANTENA = 40

# Antenas por cuadrante
antenas = {
    "Q1": (50, 50),
    "Q2": (-50, 50),
    "Q3": (-50, -50),
    "Q4": (50, -50)
}

# ----------------------------
# Generar usuarios aleatorios
# ----------------------------
usuarios = [
    (random.uniform(-LIMITE, LIMITE), random.uniform(-LIMITE, LIMITE))
    for _ in range(NUM_USUARIOS)
]

# ----------------------------
# Contar usuarios en rango
# ----------------------------
usuarios_en_rango = {nombre: 0 for nombre in antenas}

for nombre, (ax_ant, ay_ant) in antenas.items():
    for ux, uy in usuarios:
        distancia = math.sqrt((ux - ax_ant)*2 + (uy - ay_ant)*2)
        if distancia <= RADIO_ANTENA:
            usuarios_en_rango[nombre] += 1

# ----------------------------
# Mostrar resultados en consola
# ----------------------------
print("Usuarios dentro del rango de cada antena:\n")
for nombre, cantidad in usuarios_en_rango.items():
    print(f"Antenna {nombre}: {cantidad} usuarios")

# ----------------------------
# Gráfica
# ----------------------------
fig, ax = plt.subplots(figsize=(8, 8))

# Ejes
ax.axhline(0, color='black')
ax.axvline(0, color='black')

# Usuarios
x_u, y_u = zip(*usuarios)
ax.scatter(x_u, y_u, color='blue', label='Usuarios')

# Antenas y cobertura
for nombre, (x, y) in antenas.items():
    ax.scatter(x, y, color='red', s=100, marker='^')
    ax.text(x, y + 4, f"{nombre}\n({usuarios_en_rango[nombre]})", ha='center')

    circulo = plt.Circle((x, y), RADIO_ANTENA, fill=False, linestyle='--', color='red')
    ax.add_patch(circulo)

# Configuración
ax.set_xlim(-LIMITE, LIMITE)
ax.set_ylim(-LIMITE, LIMITE)
ax.set_aspect('equal')
ax.set_title("Usuarios en rango de cada antena")
ax.legend()
ax.grid(True)

plt.show()
````

Prototipo de proyecto con arduino

Este prototipo se basa en unos leds que al tener contacto con un sensor se enceinde en un intervalo de izquierda a derecha.

````python
// Pin del sensor
const int sensor = 2;

// Pines de LEDs
const int leds[] = {3, 4, 5};
const int numLeds = 3;

bool activado = false;   // Para evitar que el efecto se repita sin parar

void setup() {
  pinMode(sensor, INPUT_PULLUP);   // Activa resistencia interna

  for (int i = 0; i < numLeds; i++) {
    pinMode(leds[i], OUTPUT);
    digitalWrite(leds[i], LOW);    // Asegura que inicien apagados
  }

  Serial.begin(9600);
}

void loop() {
  int estadoSensor = digitalRead(sensor);

  // Ejecuta el efecto solo cuando se presiona / activa el sensor
  if (estadoSensor == LOW && !activado) {
    efectoViborita();
    activado = true;
  }

  // Cuando se suelta el sensor, permite que vuelva a activarse
  if (estadoSensor == HIGH) {
    activado = false;
  }
}

// Efecto de izquierda a derecha
void efectoViborita() {
  for (int i = 0; i < numLeds; i++) {
    digitalWrite(leds[i], HIGH);
    delay(70);
    digitalWrite(leds[i], LOW);
  }
}
````

FUNCIONES

Se utiliza para definir funciones, que son bloques de código reutilizables diseñados para realizar tareas específicas. Su sintaxis incluye def seguido del nombre de la función, paréntesis () para parámetros opcionales y dos puntos :, obligando a identar el cuerpo de la función.

Definición: Se utiliza una palabra clave (como def en Python o function en JS) seguida de un nombre y paréntesis para declarar la función.

Parámetros/Argumentos: Son los datos de entrada que la función necesita para operar.

Retorno (return): Es el valor final que la función entrega tras realizar su tarea.

Llamada (Call): La ejecución de la función dentro del programa principal.

Reutilización: Una vez definida, puede usarse múltiples veces en el código.

Tipos y Estructura

Funciones con retorno: Devuelven un valor calculado (ej. una suma).

Funciones sin retorno (procedimientos): Realizan una acción (ej. imprimir en pantalla) pero no devuelven datos.

Estructura: Nombre, parámetros, cuerpo de la función y valor de retorno.

````python
## FUNCION MATRIZ ALEATORIA
import random
def crearMatriz(m, n):
    A = []
    for i in range(m):
        filas = []
        for j in range(n):
            num = random.randint(0, 20)
            filas.append(num)
        A.append(filas)
    return A

## FUNCION IMPRIMIR MATRIZ
def imprimirMatriz(A):
    m=len(A)
    n=len(A[0])
    for i in range(m):
        for j in range(n):
            print(A[i][j], end="\t")
        print()

## FUNCION NUMERO MAYOR (MODULO)
def numeroMayor(A):
    mayor = A[0][0]
    indi = 0
    indj = 0

    for i in range(len(A)):
        for j in range(len(A[0])):
            if A[i][j] > mayor:
                mayor = A[i][j]
                indi = i
                indj = j

    return mayor, indi, indj

## FUNCION NUMERO MENOR (MODULO)
def numeroMenor(A):
    menor = A[0][0]
    indi = 0
    indj = 0

    for i in range(len(A)):
        for j in range(len(A[0])):
            if A[i][j] < menor:
                mayor = A[i][j]
                indi = i
                indj = j

    return menor, indi, indj

# FUNCION SUMA DE TODOS LOS NUMEROS
def sumaMatriz(A):
    suma = 0
    for i in range(len(A)):
        for j in range(len(A[0])):
            suma += A[i][j]
    return suma

## FUNCION NUMEROS PARES
def numerosPares(A):
    pares = []
    for i in range(len(A)):
        for j in range(len(A[0])):
            if A[i][j] % 2 == 0:
                pares.append(A[i][j])
    return pares

## FUNCION NUMEROS IMPARES
def numerosImpares(A):
    impares = []
    for i in range(len(A)):
        for j in range(len(A[0])):
            if A[i][j] % 2 != 0:
                impares.append(A[i][j])
    return impares

## FUNCION SUMA DE FILAS
def sumaFilas(A):
    vectorSuma = []
    for i in range(len(A)):
        suma = 0
        for j in range(len(A[0])):
            suma += A[i][j]
        vectorSuma.append(suma)
    return vectorSuma

## FUNCION SUMA DE COLUMNAS
def sumaColumnas(A):
    n = len(A[0])
    vectorSuma = [0] * n

    for i in range(len(A)):
        for j in range(n):
            vectorSuma[j] += A[i][j]

    return vectorSuma

## FUNCION MAYOR DE CADA FILA
def mayorPorFila(A):
    vectorMayor = []

    for i in range(len(A)):
        mayor = A[i][0]
        for j in range(len(A[0])):
            if A[i][j] > mayor:
                mayor = A[i][j]
        vectorMayor.append(mayor)

    return vectorMayor

## FUNCION MAYOR DE CADA COLUMNA
def mayorPorColumna(A):
    n = len(A[0])
    vectorMayor = []

    for j in range(n):
        mayor = A[0][j]
        for i in range(len(A)):
            if A[i][j] > mayor:
                mayor = A[i][j]
        vectorMayor.append(mayor)

    return vectorMayor

## FUNCION SUMA DIAGONAL PRINCIPAL
def sumaDiagonalPrincipal(A):
    suma = 0
    for i in range(len(A)):
        suma += A[i][i]
    return suma

## FUNCION SUMA DIAGONAL SECUNDARIA
def sumaDiagonalSecundaria(A):
    n = len(A)
    suma = 0
    for i in range(n):
        suma += A[i][n - 1 - i]
    return suma


# PROGRAMA PRINCIPAL
print("Primera matriz")
A = crearMatriz(4, 4)
imprimirMatriz(A)

mayor, fila, columna = numeroMayor(A)
print("El número mayor es:", mayor)
print("Está en la posición: fila", fila, "columna", columna)

menor, fila, columna = numeroMenor(A)
print("El número menor es:", menor)
print("Está en la posición: fila", fila, "columna", columna)

totalsuma = sumaMatriz(A)
print("La suma es:", totalsuma)

pares = numerosPares(A)
print("Números pares encontrados:", pares)

impares = numerosImpares(A)
print("Números impares encontrados:", impares)

suma_filas = sumaFilas(A)
print("Suma de cada fila:", suma_filas)

suma_columnas = sumaColumnas(A)
print("Suma de cada columna:", suma_columnas)

mayores_fila = mayorPorFila(A)
print("Mayor de cada fila:", mayores_fila)

mayores_columna = mayorPorColumna(A)
print("Mayor de cada columna:", mayores_columna)

suma_diag = sumaDiagonalPrincipal(A)
print("Suma de la diagonal principal:", suma_diag)

suma_diag_sec = sumaDiagonalSecundaria(A)
print("Suma de la diagonal secundaria:", suma_diag_sec)
````

**Ejercicio 7.12**
 Algoritmo que triangule una matriz cuadrada y halle su determinante. En las matrices cuadradas el valor del determinante coincide con el producto de los elementos de la diagonal de la matriz triangulada, multiplicado por –1
tantas veces como hayamos intercambiado filas al triangular la matriz.
Proceso de triangulación de una matriz para todo i desde 1 hasta n – 1 hacer:
a) Si el elemento de lugar (i,i) es nulo, intercambiar filas hasta que dicho elemento sea no nulo o agotar los
posibles intercambios.
b) A continuación se busca el primer elemento no nulo de la fila i-ésima y, en el caso de existir, se usa para
hacer ceros en la columna de abajo.
Estructuras de datos I (arrays y estructuras) 281
 Sea dicho elemento matriz[i,r]
 Multiplicar fila i por matriz[i+1,r]/matriz[i,r] y restarlo a la i+1
 Multiplicar fila i por matriz[i+2,r]/matriz[i,r] y restarlo a la i+2
 ............................................................................................................
 Multiplicar fila i por matriz[m,r]/matriz[i,r] y restarlo a la m.
```python
n = int(input("Tamaño de la matriz cuadrada: "))

# Cargar matriz
matriz = []
for i in range(n):
    fila = []
    for j in range(n):
        fila.append(float(input(f"Elemento [{i+1}][{j+1}]: ")))
    matriz.append(fila)

determinante = 1
intercambios = 0

# Eliminación Gaussiana
for i in range(n):
    # Si el pivote es 0, buscar intercambio
    if matriz[i][i] == 0:
        for k in range(i+1, n):
            if matriz[k][i] != 0:
                matriz[i], matriz[k] = matriz[k], matriz[i]
                intercambios += 1
                break

    pivote = matriz[i][i]

    if pivote != 0:
        for j in range(i+1, n):
            factor = matriz[j][i] / pivote
            for k in range(i, n):
                matriz[j][k] -= factor * matriz[i][k]

# Producto diagonal
for i in range(n):
    determinante *= matriz[i][i]

# Ajustar signo por intercambios
if intercambios % 2 != 0:
    determinante *= -1

print("Matriz triangular:")
for fila in matriz:
    print(fila)

print("Determinante:", determinante)
```

6.2. Diseñar un algoritmo que calcule el máximo común divisor de dos números mediante el algoritmo de Euclides.
Sean los dos números A y B. El método para hallar el máximo común divisor (mcd) de dos números A y B por el método
de Euclides es:
1. Dividir el número mayor (A) por el menor (B). Si el resto de la división es cero, el número B es el máximo común
divisor.
2. Si la división no es exacta, se divide el número menor (B) por el resto de la división anterior.
3. Se siguen los pasos anteriores hasta obtener un resto cero. El último divisor es el mcd buscado.

**Python**

````python
def mcd(a, b):
    while a != b:
        if a > b:
            a = a - b
        else:
            b = b - a
    return a
num1 = int(input("Ingrese el primer número: "))
num2 = int(input("Ingrese el segundo número: "))
resultado = mcd(num1, num2)
print("El máximo común divisor es:", resultado)
````

````python
# Pedir datos al usuario
dividendo = int(input("Ingrese el dividendo: "))
divisor = int(input("Ingrese el divisor: "))

# Inicializar cociente y acumulador
cociente = 0
acumulador = dividendo

# Restar el divisor repetidamente hasta que el acumulador sea menor que divisor
while acumulador >= divisor:
    acumulador -= divisor  # equivalente a acumulador = acumulador - divisor
    cociente += 1          # aumentar cociente en 1

# Lo que queda en acumulador es el resto
resto = acumulador

# Mostrar resultados
print("Cociente:", cociente)
print("Resto:", resto)
````

6.12. Escribir una función que permita deducir si una fecha
leída del teclado es válida.

**Phyton**

````python
# Función que verifica si una fecha es válida
def es_fecha_valida(dia, mes, anio):
    # Comprobar mes válido
    if mes < 1 or mes > 12:
        return False

    # Determinar número de días según el mes
    if mes in [1, 3, 5, 7, 8, 10, 12]:
        dias_mes = 31
    elif mes in [4, 6, 9, 11]:
        dias_mes = 30
    else:  # Febrero
        # Año bisiesto: divisible por 4 y (no divisible por 100 o divisible por 400)
        if (anio % 4 == 0) and ((anio % 100 != 0) or (anio % 400 == 0)):
            dias_mes = 29
        else:
            dias_mes = 28

    # Comprobar día válido
    if dia < 1 or dia > dias_mes:
        return False

    return True

# Pedir fecha al usuario
dia = int(input("Ingrese el día: "))
mes = int(input("Ingrese el mes: "))
anio = int(input("Ingrese el año: "))

# Validar la fecha usando la función
if es_fecha_valida(dia, mes, anio):
    print("La fecha es válida.")
else:
    print("La fecha NO es válida.")
````

calcule el factorial de un número entero positivo, b) la
potencia de un número entero positivo.

**Phyton**

````python
# factorial recursivo
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)


# potencia recursiva
def potencia(base, exponente):
    if exponente == 0:
        return 1
    else:
        return base * potencia(base, exponente - 1)


# programa principal
n = int(input("Ingrese un numero para calcular el factorial: "))
print("El factorial es:", factorial(n))

b = int(input("Ingrese la base: "))
e = int(input("Ingrese el exponente: "))
print("La potencia es:", potencia(b, e))
````

**Pseint**

````python
import math

# Función para leer la lista de números
def leer_array():
    while True:
        n = int(input("Diga el número de elementos de la lista (max 15): "))
        if n <= 15:
            break
        print("Error: el número máximo de elementos es 15.")

    x = []
    print("Deme los elementos:")
    for i in range(n):
        valor = float(input(f"x[{i+1}]: "))
        x.append(valor)
    return x, n

# Función para calcular la desviación estándar
def desviacion(x, n):
    suma = sum(x)
    xm = suma / n  # media
    sigma = 0
    for i in range(n):
        sigma += (x[i] - xm) ** 2
    # Desviación estándar muestral
    return math.sqrt(sigma / (n - 1))

# Programa principal
x, n = leer_array()
sigma = desviacion(x, n)
print(f"\nLa desviación estándar es: {sigma}")
````

````python
def dibujar_cuadrado(n):
    # Crear matriz n x n inicializada con ceros
    a = [[0] * n for _ in range(n)]

    # Posición inicial del número 1
    i = 0           # primera fila (Python index empieza en 0)
    j = n // 2      # columna central

    # Llenar el cuadrado mágico
    for c in range(1, n * n + 1):
        a[i][j] = c

        # Guardar la posición actual
        i_anterior, j_anterior = i, j

        # Próxima posición: arriba y a la derecha
        i -= 1
        j += 1

        # Ajuste para ciclicidad
        if i < 0:
            i = n - 1
        if j >= n:
            j = 0

        # Si la casilla ya está ocupada
        if a[i][j] != 0:
            i = i_anterior + 1
            j = j_anterior
            if i >= n:
                i = 0

    # Imprimir el cuadrado
    for fila in a:
        for num in fila:
            print(f"{num:3}", end=" ")  # ancho de 3 para alinear
        print()  # nueva línea por fila

# Programa principal
while True:
    n = int(input("Dame las dimensiones del cuadrado (impar, entre 3 y 11): "))
    if n % 2 != 0 and 3 <= n <= 11:
        break
    print("¡Error! Debe ser un número impar entre 3 y 11.")

dibujar_cuadrado(n)
````

7.12. Algoritmo que triangule una matriz cuadrada y halle su determinante. En las matrices cuadradas el valor del de
terminante coincide con el producto de los elementos de la diagonal de la matriz triangulada, multiplicado por –1
tantas veces como hayamos intercambiado filas al triangular la matriz.
Proceso de triangulación de una matriz para todo i desde 1 hasta n – 1 hacer:
a)  Si el elemento de lugar (i,i) es nulo, intercambiar filas hasta que dicho elemento sea no nulo o agotar los
posibles intercambios.
b)  
A continuación se busca el primer elemento no nulo de la fila i-ésima y, en el caso de existir, se usa para
hacer ceros en la columna de abajo.
Estructuras de datos I  (arrays y estructuras)  
281
Sea dicho elemento matriz[i,r]
Multiplicar fila i por matriz[i+1,r]/matriz[i,r] y restarlo a la i+1
Multiplicar fila i por matriz[i+2,r]/matriz[i,r] y restarlo a la i+2
 ............................................................................................................
Multiplicar fila i por matriz[m,r]/matriz[i,r] y restarlo a la m

````python
def leer_matriz(n):
    """Lee una matriz cuadrada de tamaño n x n"""
    matriz = []
    print("Deme los valores para la matriz:")
    for i in range(n):
        fila = []
        for j in range(n):
            val = float(input(f"M[{i+1},{j+1}]: "))
            fila.append(val)
        matriz.append(fila)
    return matriz

def escribir_matriz(matriz):
    """Imprime la matriz en pantalla"""
    print("Matriz triangulada:")
    for fila in matriz:
        for val in fila:
            print(f"{val:10.4f}", end=" ")
        print()
    input("Pulse Enter para continuar...")

def intercambiar_filas(matriz, i, j):
    """Intercambia las filas i y j de la matriz"""
    matriz[i], matriz[j] = matriz[j], matriz[i]

def triangula(matriz):
    """Triangula la matriz y devuelve su determinante"""
    n = len(matriz)
    signo = 1

    for i in range(n-1):
        # Si el elemento pivote es cero, buscar fila con valor no nulo y hacer intercambio
        if matriz[i][i] == 0:
            t = 1
            while i + t < n and matriz[i+t][i] == 0:
                t += 1
            if i + t < n:
                intercambiar_filas(matriz, i, i+t)
                signo *= -1
                escribir_matriz(matriz)

        # Buscar primer elemento no nulo en la fila i desde la columna i en adelante
        r = i
        while r < n and matriz[i][r] == 0:
            r += 1

        if r < n and matriz[i][r] != 0:
            # Hacer ceros debajo del pivote
            for t in range(i+1, n):
                if matriz[t][r] != 0:
                    factor = matriz[t][r] / matriz[i][r]
                    for j in range(r, n):
                        matriz[t][j] -= matriz[i][j] * factor
                    escribir_matriz(matriz)

    # Calcular determinante como producto de la diagonal principal
    dt = signo
    for i in range(n):
        dt *= matriz[i][i]
    return dt

# -------------------------------
# Programa principal
n = int(input("Ingrese el tamaño de la matriz cuadrada (n x n): "))
matriz = leer_matriz(n)
determinante = triangula(matriz)
print(f"Determinante = {determinante:.4f}")
````
